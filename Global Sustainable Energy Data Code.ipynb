```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Code\\Data Analysis\\Datasets\\Kaggle Datasets\\global-data-on-sustainable-energy.csv")
```

```python
#Replacing missing values
df = df.fillna({'Access to clean fuels for cooking' : 0,'Renewable-electricity-generating-capacity-per-capita' : 0,
                'Financial flows to developing countries (US $)' : 0, 'Renewable energy share in the total final energy consumption (%)' : 0,
                'Electricity from fossil fuels (TWh)' : 0, 'Electricity from nuclear (TWh)' : 0,
          'Electricity from renewables (TWh)' : 0, 'Low-carbon electricity (% electricity)' : 0,  'Primary energy consumption per capita (kWh/person)' : 0,
                'Energy intensity level of primary energy (MJ/$2017 PPP GDP)' : 0, 'Value_co2_emissions_kt_by_country' : 0,
          'Renewables (% equivalent primary energy)' : 0, 'gdp_growth' : 0, 'gdp_per_capita' : 0, 'Density\n(P/Km2)' : 0, 'land_area' : 0, 'Latitude' : 0, 'Longitude' : 0})
```

```python
#Replacing missing values in the Access to electricity column using the backfill method
df['Access to electricity (% of population)'] = df['Access to electricity (% of population)'].fillna(method = 'bfill')
```

```python
#Renaming the Entity column to Country
df.rename(columns = {'Entity' : 'Country'}, inplace=True)
```

```python
df = round(df, 2)
```

```python
df.head()
```

```python
#World electricity statistics by year
electricity_stats = df.groupby('Year')['Access to electricity (% of population)'].describe()
```

```python
electricity_stats
```

```python
# Electricity access by country from 2000 to 2020
avg_electricity_by_country = df.groupby(['Year', 'Country'])['Access to electricity (% of population)'].mean().unstack()
avg_electricity_by_country
```

```python
mean_median_access_to_electricity = round(df.groupby('Year')['Access to electricity (% of population)'].agg(['mean', 'median']), 2)
```

```python
mean_median_access_to_electricity
```

```python
# World electricty access trend from 2000 to 2020
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x='Year', y='Access to electricity (% of population)', errorbar=None)
plt.title('Global Access to Electricity 2000 to 2020')
plt.ylabel('Access to Electricity (% of Population)')
plt.xlabel('Year')
plt.savefig('Global Access to Electricity 2000 to 2020.jpg')
plt.show()
```

```python
# 10 countries with the least access to electricity in 2000
electricity_data_2000 = df.query('Year == 2000')
least_electricity_access_countries_2000 = electricity_data_2000[['Country', 'Access to electricity (% of population)']].nsmallest(10, 'Access to electricity (% of population)').reset_index(drop = True)
least_electricity_access_countries_2000
```

```python
sns.barplot(x = 'Access to electricity (% of population)', y = 'Country', data = least_electricity_access_countries_2000)
plt.title('10 countries with least access to electricity in 2000')
plt.xlabel('Access to electricity (% of population')
plt.ylabel('Countries')
plt.savefig('10 countries with least access to electricity in 2000.jpg')
plt.show()
```

```python
# 10 Countries with the least access to electricity in 2020
electricity_data_2020 = df.query('Year == 2020')
least_electricity_access_countries_2020 = electricity_data_2020[['Country', 'Access to electricity (% of population)']].nsmallest(10, 'Access to electricity (% of population)').reset_index(drop = True)
least_electricity_access_countries_2020
```

```python
sns.barplot(x = 'Access to electricity (% of population)', y = 'Country', data = least_electricity_access_countries_2020)
plt.title('10 countries with least access to electricity in 2020')
plt.xlabel('Access to electricity (% of population')
plt.ylabel('Countries')
plt.savefig('10 countries with least access to electricity in 2020.jpg')
plt.show()
```
